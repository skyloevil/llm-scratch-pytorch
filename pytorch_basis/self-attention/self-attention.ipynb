```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SelfAttention(nn.Module):
    def __init__(self,embed_size):
        super(SelfAttention,self).__init__()
        self.embed_size = embed_size
        # çº¿æ€§å±‚ï¼Œç”¨äºç”ŸæˆæŸ¥è¯¢ã€é”®å’Œå€¼å‘é‡
        # y = Wx + b
        self.query = nn.Linear(embed_size,embed_size,bias=False)
        self.key = nn.Linear(embed_size,embed_size,bias=False)
        self.value = nn.Linear(embed_size,embed_size,bias=False)
    
    def forward(self,datas):
        Q = self.query(datas)
        K = self.key(datas)
        V = self.value(datas)

        # Q: (batch_size, seq_len, embed_size)
        # KT: (batch_size, embed_size, seq_len)
        P = torch.matmul(Q,K.transpose(-1,-2)) 
        # P: (batch_size, seq_len, seq_len)
        attention_scores = P/torch.sqrt(torch.tensor(self.embed_size))
        print("attention_scores:\n",attention_scores.shape,"\n")
        # å½’ä¸€åŒ–
        attention_weight = F.softmax(attention_scores, dim=-1)
        output = torch.matmul(attention_weight,V)
        # output: (batch_size, seq_len, embed_size)


        return output
```

```python
datas = torch.rand(1,2,4)
net = SelfAttention(4)
output = net(datas)
print(output.shape)
```
